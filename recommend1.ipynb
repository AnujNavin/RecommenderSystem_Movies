```python
#importing all the necessary built-in libraries that we gonna need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#reading our csv file from where it is saved in computer
df = pd.read_csv(r"C:\Users\Anuj Navin\Downloads\7181_10279_compressed_movie_metadata.csv\movie_metadata.csv")
```

```python
#displays total number of NaN values in each coulmn(feature)
df.isnull().sum()
```

```python
#handling NaN values,simply filling them with unknown and doing other data processing
df['actor_1_name']  = df['actor_1_name'].fillna('unknown')
df['actor_2_name']  = df['actor_2_name'].fillna('unknown')
df['actor_3_name']  = df['actor_3_name'].fillna('unknown')
df['director_name'] = df['director_name'].fillna('unknown')
df['genres']  = df['genres'].str.replace('|',' ')
df['movie_title'] =  df['movie_title'].str.lower()
```

```python
#removing duplicate movie titles
df = df.drop_duplicates(['movie_title'])
df.shape
```

```python
#most important step for content based recommender system making a combined column out of director name,actors name,genre etc.
df['comb'] =  df['actor_1_name'] + df['actor_2_name'] + df['actor_2_name'] + df['genres'] + df['director_name']
```

```python
#creating an instance of tf-idf vectorizer
cv =  TfidfVectorizer()
```

```python
df1 = cv.fit_transform(df['comb'])
```

```python
df1
```

```python
#calling cosine similiarity here which we use to find how close vectors generated by tf-idf are
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df1 = pd.DataFrame(cosine_similarity(df1))
df1
 
```

```python
df1.index = df['movie_title']
```
